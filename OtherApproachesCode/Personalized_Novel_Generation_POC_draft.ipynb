````python
def get_llm_response(prompt, temp):
    chat_completion = [
        {
            "role": "user",
            "content": [{"text": prompt}],
        }
    ]
    response = client.converse(
        modelId=model_id,
        messages=chat_completion,
        inferenceConfig={"maxTokens": 4096, "temperature": temp, "topP": 0.8},
    )
    response_data = {"prompt": prompt, "response": response["output"]["message"]["content"][0]["text"]}
    # print("Prompt: " + prompt)
    # print("LLM Response: " + response_data['response'])
    return response_data['response']
````

````python
def get_llm_response_with_token(prompt, temp, token_prefix, token_suffix):
    response_actual_text = get_llm_response(prompt, temp)
    start_index = response_actual_text.find(token_prefix)
    end_index = response_actual_text.find(token_suffix)
    personalized_output = response_actual_text[start_index + len(token_prefix) : end_index].strip()
    return personalized_output
````
